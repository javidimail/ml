```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import sys
```

```python
df_housing = pd.read_csv('cal_housing_clean.csv')
df_housing.head()
```

```python
df_housing.info()
```

```python
sys.executable
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df_housing, test_size = 0.3, random_state=202)
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
y_test = df_test['medianHouseValue']
x_test = df_test.drop('medianHouseValue',axis=1)
y_train = df_train['medianHouseValue']
x_train = df_train.drop('medianHouseValue', axis=1)
```

```python
type(y_test)
```

```python
type(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train
```

```python
#Compute the minimum and maximum to be used for later scaling.
scaler.fit(x_train)
```

```python
# Scaling features of X according to feature_range and then recreate the data frame
x_train = pd.DataFrame(data=scaler.transform(x_train), 
                       columns=x_train.columns,
                       index=x_train.index)
```

```python
x_train
```

```python
x_test = pd.DataFrame(data=scaler.transform(x_test), 
                       columns=x_test.columns,
                       index=x_test.index)
```

```python
x_test
```

```python
df_housing.columns
```

```python
housingMedianAge  = tf.feature_column.numeric_column('housingMedianAge')
totalRooms  = tf.feature_column.numeric_column('totalRooms')
totalBedrooms  = tf.feature_column.numeric_column('totalBedrooms')
population  = tf.feature_column.numeric_column('population')
households  = tf.feature_column.numeric_column('households')
medianIncome  = tf.feature_column.numeric_column('medianIncome')
```

```python
feat_cols = [housingMedianAge,totalRooms,
totalBedrooms,
population,
households,
medianIncome]
housingMedianAge
```

```python
fn_input = tf.estimator.inputs.pandas_input_fn(x=x_train, 
                                               y=y_train,
                                              batch_size = 100,
                                              shuffle = True)
```

```python
dnn_model = tf.estimator.DNNRegressor(hidden_units=[6,6,6], 
                                      feature_columns=feat_cols)
```

```python
dnn_model.train(input_fn=fn_input, steps=20000)
```

```python
fn_predict = tf.estimator.inputs.pandas_input_fn(x=x_test,
                                                batch_size=10,
                                                num_epochs=1,
                                                shuffle=False)
```

```python
predict_results = dnn_model.predict(fn_predict)
```

```python
predict_list = list(predict_results)
predict_list
```

```python
predict_list_final = []
for predict_item in predict_list:
    predict_list_final.append(predict_item['predictions'])
```

```python
predict_list_final
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, predict_list_final)**0.2
```
