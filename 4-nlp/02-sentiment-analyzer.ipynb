```python
wordnet_lemmatizer = WordNetLemmatizer()
stopwords = set(w.rstrip() for w in open('datasets/stopwords.txt'))
```

```python
word_index_map = {}
current_index  = 0

positive_tokenized = []
negative_tokenized = []

nltk.download('punkt')
nltk.download("wordnet")
def my_tokenizer(s):
    s = s.lower()
    tokens = nltk.tokenize.word_tokenize(s)
    tokens = [t for t in tokens if len(t) > 2]
    tokens = [wordnet_lemmatizer.lemmatize(t) for t in tokens]
    tokens = [t for t in tokens if t not in stopwords]
    print tokens
    return tokens

for review in positive_reviews:
    print review.text
    tokens = my_tokenizer(review.text)
    positive_tokenized.append(tokens)
    for token in tokens:
        if token not in word_index_map:
            word_index_map[token] = current_index
            current_index +=1
```

```python
word_index_map
```

```python
for review in negative_reviews:
    tokens = my_tokenizer(review.text)
    negative_tokenized.append(tokens)
    for token in tokens:
        if token not in word_index_map:
            word_index_map[token] = current_index
            current_index +=1
```

```python
len(word_index_map)
```

```python
def tokens_to_vector(tokens, label):
    x = np.zeros(len(word_index_map) + 1)
    for t in tokens:
        i = word_index_map[t]
        x[i] += 1
    x = x / x.sum()
    x[-1] = label
    return x
    
```

```python
N = len(positive_tokenized) + len(negative_tokenized)
data = np.zeros((N, len(word_index_map)+1))
i = 0
for tokens in positive_tokenized:
    xy = tokens_to_vector(tokens,1) # 1 is the label, postitive reviews
    data[i,:] = xy
    i += 1
for tokens in negative_tokenized:
    xy = tokens_to_vector(tokens,0) # 0 is the label, postitive reviews
    data[i,:] = xy
    i += 1
```

```python
np.random.shuffle(data)
X = data[:,:-1]
Y = data[:, -1]
```

```python
Xtrain = X[:-100,]
Ytrain = Y[:-100,]
Xtest = X[-100:,]
Ytest = Y[-100:,]
```

```python
model = LogisticRegression()
```

```python
model.fit(Xtrain, Ytrain)
print "classifiaction rate: ", model.score(Xtest,Ytest)
```

```python
treshold = 0.5
for word, index in word_index_map.iteritems():
    weight = model.coef_[0][index]
    if weight > treshold or weight < -treshold:
        print word, weight
```
